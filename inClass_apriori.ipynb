```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
oht = TransactionEncoder() 
oht_array = oht.fit(dataset).transform(dataset) 
```

```python
oht_array
```

```python
oht_df = pd.DataFrame(oht_array, columns = oht.columns_) 
```

```python
oht_df.head()
```

```python
oht_df.tail()
```

```python
oht_df.drop(oht_df.columns[0], axis = 1)
```

```python
frequent_items = apriori(oht_df, min_support = 0.6, use_colnames = True)
print(frequent_items)
```

```python
type(frequent_items)
```

```python
frequent_items.sort_values(by = ['support'], ascending = False)
```

```python
rules = association_rules(frequent_items, metric = 'confidence', min_threshold = 0.7) 
print(print(rules[['antecedents', 'consequents', 'support', 'confidence']]))
```

```python
rules.sort_values(by = ['confidence'], ascending = False)
```
